```python
#Data Cleaning and Transformation
```

```python
# Handle Missing Values
# Fill numeric missing values with mean, and categorical with mode
num_cols = df.select_dtypes(include=['float64', 'int64']).columns
cat_cols = df.select_dtypes(include=['object']).columns

df[num_cols] = df[num_cols].apply(lambda x: x.fillna(x.mean()))
df[cat_cols] = df[cat_cols].apply(lambda x: x.fillna(x.mode()[0]))
```

```python
# Fix Data Type Mismatches
# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y', errors='coerce')
```

```python
# Remove Duplicates (if any)
df = df.drop_duplicates()
```

```python
# Reset Index after cleaning
df.reset_index(drop=True, inplace=True)
```

```python
# Show cleaned dataset summary
print("\n----- Cleaned Dataset Info -----")
df.info()
print("\n----- Missing Values After Cleaning -----")
print(df.isnull().sum())
```

```python
# Preview cleaned data
print("\n----- Cleaned Data Preview -----")
display(df.head())
```

```python
# Display all columns
pd.set_option('display.max_columns', None)

# Load the dataset (from your SQLite or CSV)
df = pd.read_csv("marketing_campaign_dataset.csv")

# Check structure
print("Shape:", df.shape)
df.head()
```

```python
# Top campaign based on engagement score 

Top_Engagement_Campaigns = df[df['Engagement_Score'] > 7]
Top_Engagement_Campaigns
```

```python
#Campaign with the Maximum Click Engagement
Highest_click_campaign = df['Clicks'].idxmax()
print(f'Maximum number of clicks per campaign:',Highest_click_campaign)
```

```python
# Highest conversion rate

Highest_conv_rate = df['Conversion_Rate'].idxmax()
print(f'Highest conversion rate is:',Highest_conv_rate)
```

```python
# Most successful campaign of the september month

df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

september_data = df[df['Date'].dt.month == 9]

Most_successful_campaign_in_sep = september_data.loc[september_data['Acquisition_Cost'].idxmax()]

print('\n Most successful campaign of september \n', Most_successful_campaign_in_sep)
```

```python
# Count of campaign per month

df['Date'] = pd.to_datetime(df['Date'])

df['Month'] = df['Date'].dt.to_period('M')

campaign_per_month = df.groupby('Month')['Campaign_ID'].count()
print('\n Campaign count per month \n',campaign_per_month)
```

```python
# Highest number of campaigns per month

Highest_campaign_per_month = df.groupby('Month')['Campaign_ID'].count().idxmax()
print('Month which has highest campaign:',Highest_campaign_per_month)
```

```python
# Highest num of clicks based on channel

Highest_click = df.groupby('Channel_Used')['Clicks'].sum().sort_values(ascending= False).idxmax()
print(f'Highest clicks based on channed used:',Highest_click)
```

```python
# Higly active location

Highly_active_location = df.groupby('Location')['Clicks'].mean().idxmax()
print(f'Highly active location is:',Highly_active_location)
```

```python
# Channel prefered by the customers

channel_perference = df.groupby("Channel_Used").agg({
    "Conversion_Rate": "mean",
    "ROI": "mean",
    "Acquisition_Cost": "mean"
}).reset_index().round(2)

print("\n Channel Effectiveness: \n", channel_perference)
```

```python
# Location preference by the customers

location_perference = df.groupby("Location").agg({
    "Acquisition_Cost": "mean",
    "Conversion_Rate": "mean",
    "ROI": "mean"
}).reset_index().round(2)

print("\n Location Performance:\n", location_perference)
```

```python
# Type of campaign prefered by the customers 

type_perf = df.groupby("Campaign_Type").agg({
    "ROI": "mean",
    "Conversion_Rate": "mean",
    "Acquisition_Cost": "mean"
}).reset_index().round(2)

print("\ Campaign Type Performance:\n", type_perf)
```

```python
Top_segment = df.groupby('Customer_Segment')['Engagement_Score'].mean().idxmax()
Top_segment
```

```python
# Language preference by the customers

lang_perf = df.groupby("Language").agg({
    "ROI": "mean",
    "Engagement_Score": "mean"
}).reset_index().round(2)

print("\n Language Effectiveness:\n", lang_perf)
```

```python
# Clicks vs conversion correlation

click_conversion_corr = df["Clicks"].corr(df["Conversion_Rate"])
print("\n Correlation between Clicks and Conversion Rate:", click_conversion_corr)

plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="Clicks", y="Conversion_Rate", color='skyblue', s=70, edgecolor='black')
plt.title("Clicks vs Conversion Rate")
plt.xlabel("Clicks")
plt.ylabel("Conversion Rate")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

```python
# impression_engagement correlation

impression_engagement_corr = df["Impressions"].corr(df["Engagement_Score"])
print("\n Correlation between Impressions and Engagement Score:", impression_engagement_corr)
```

```python
# Cost efficiency per campaign

df["Cost_per_Conversion"] = df["Acquisition_Cost"] / df["Conversion_Rate"].replace(0, pd.NA)
cost_eff = df[["Campaign_ID", "Company", "Cost_per_Conversion"]].round(2)

print("\n Cost Efficiency (per campaign):\n", cost_eff)
```

```python
# Time preference by the customers

time_perf = df.groupby("Duration").agg({
    "ROI": "mean",
    "Conversion_Rate": "mean"
}).reset_index().round(2)

print("\n Duration vs Performance:\n", time_perf)
```

```python
#Central tendencies of the ROI
Mean_of_ROI = df['ROI'].mean().round(2)
print(f"Mean of ROI is :",Mean_of_ROI)

Median_of_ROI = df['ROI'].median()
print(f"Median of ROI is :",Median_of_ROI)

Mode_of_ROI = df['ROI'].mode()[0]
print(f"Mode of ROI is :",Mode_of_ROI)

range_of_ROI=df['ROI'].max() - df['ROI'].min()
print(f"Range of ROI is :",range_of_ROI)

Var_of_ROI = df['ROI'].var()
print(f"Variance of ROI is :",Var_of_ROI)

Std_of_ROI = df['ROI'].std()
print(f"Standard Deviation of ROI is :",Std_of_ROI)
```

```python
#Skewness and Kustosis analysis of Conversion_Rate
from scipy.stats import skew, kurtosis
skewness = skew(df['Conversion_Rate'])
kurt = kurtosis(df['Conversion_Rate'])
print(f"skewness:",skewness)
print(f"Kurtosis:",kurt)
```

```python
# --- Distribution of ROI ---

plt.figure(figsize=(8,5))
sns.histplot(df['ROI'], kde=True, color='skyblue', bins=20)
plt.title("Distribution of ROI")
plt.xlabel("ROI")
plt.ylabel("Frequency")
plt.show()
print("""
Insight:
1.Uniform Distribution: ROI values appear evenly spread between 2 and 8, suggesting campaigns have a wide and balanced performance range.

2.No Major Outliers: The smooth KDE curve shows no extreme peaks — ROI values are fairly consistent.

3.Stable Campaign Returns: Most campaigns deliver similar ROI levels, indicating predictable performance across campaigns.
""")
```

```python
# --- Distribution of Conversion Rate ---
plt.figure(figsize=(8,5))
sns.boxplot(x=df['Conversion_Rate'], color='lightgreen')
plt.title("Conversion Rate Spread")
plt.show()
print("""
Insight:
1.Median near center of box → Almost Symmetric:The median line is roughly in the middle of the box, showing that the data is not heavily skewed.

2.Equal whisker lengths on both sides:Both whiskers (lines extending from the box) are nearly equal — indicating balanced spread of data on both sides.

3.No outliers visible:Suggests no extreme low or high conversion rates, hence the distribution is normal or slightly uniform.
""")
```

```python
# --- Correlation Heatmap for Numerical Variables ---
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix (Numerical Features)")
plt.show()
print("""
Insight:
1.Very weak correlations: All numerical features show correlations close to 0, meaning there are no strong linear relationships between engagement, clicks, impressions, conversion rate, ROI, or acquisition cost.

2.Independence of Engagement Score: Engagement Score does not strongly correlate with any other metric, suggesting it may be influenced by qualitative factors (e.g., content quality, targeting) rather than just campaign metrics.
""")
```

```python
# --- Pivot Table: ROI by Company and Channel Used ---
pivot_roi = df.pivot_table(values='ROI', index='Company', columns='Channel_Used', aggfunc='mean')
print(pivot_roi)
plt.figure(figsize=(10,6))
sns.heatmap(pivot_roi, annot=True, cmap='YlGnBu')
plt.title("Average ROI by Company and Channel Used")
plt.show()
print("""
Insight:
1.ROI values are very close (around 5) across all companies and channels, indicating minimal variation in performance.

2.A few slight peaks are visible (e.g., DataTech Solutions–Website and Alpha Innovations–YouTube), but differences are marginal.

3.This suggests that ROI is relatively stable across both companies and channels, meaning channel choice may not strongly impact ROI in this dataset.
""")
```

```python
# --- Relationship among Clicks, Impressions, and ROI ---
plt.figure(figsize=(8,5))
sns.scatterplot(x='Impressions', y='ROI', size='Clicks', hue='Company', data=df, alpha=0.7)
plt.title("ROI vs Impressions (Bubble size = Clicks)")
plt.show()
print("""
Insight:
1.Positive Correlation (Impressions vs ROI):
ROI generally increases as Impressions increase, indicating that higher visibility contributes to better returns.

2.Click Engagement Impact:
Larger bubbles (more clicks) often cluster in higher ROI zones — showing that campaigns with more user engagement tend to yield stronger ROI.

3.Best 3 performing companies in terms of ROI efficiency and engagement (Clicks per Impressions) appear to be:
    Alpha Innovations (green):Consistently achieves high ROI even with moderate impressions.
    DataTech Solutions (red):Shows stable ROI growth with impressions.
    Innovate Industries (blue):Maintains balanced ROI across a wide range of impressions.
""")
```

```python
#Exploring Interactions Between ROI, Conversion Efficiency, and Click Performance
# Select the numerical features
X = df[['ROI', 'Conversion_Rate', 'Clicks']]
# Create pairplot
sns.pairplot(
    X,
    plot_kws={'c': 'orange', 'alpha': 0.6}  # sets color and transparency
)
# Add a super title
plt.suptitle('Pairwise Relationships of ROI, Conversion Rate, and Clicks', y=1.02)
# Display the plot
plt.show()
print("""
Insight:
1.ROI shows a mild positive trend with Clicks, indicating engagement supports better returns.

2.Conversion Rate remains fairly constant, showing limited impact on ROI.

3.Overall, no strong linear correlation exists among the three metrics.
""")
```
